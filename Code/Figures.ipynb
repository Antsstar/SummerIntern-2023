```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_pickle("Results(p=0.1).pkl") #Loads
df
```

```python
#d is an array
# for Histograms




#d = df.loc[df.DL_Mode=='8PSK'].DL_capacity.to_numpy()
d = df.loc[df.Ground_Station=='London'].DL_capacity.to_numpy()

 

fig, ax = plt.subplots()
n, bins, patches = plt.hist(x=d, bins='auto', color='#0504aa',
                            alpha=0.7, rwidth=0.85)
plt.grid(axis='y', alpha=0.75)
#plt.xlabel('SINR (dB)')
plt.xlabel('Data Rate (Mbps)')
plt.ylabel('# of Occurence')
#plt.title(f'{link_type} Histogram of {loaded_case} SINR \n Margin {margin_input}')

link_type = 'Downlink'
loaded_case = 'London'
margin_input = 'p = 0.1'
plt.title(f'{link_type} Histogram of {loaded_case} Capacity \n Margin {margin_input}, Mean = {np.round(d.mean())} (Mbps)')

 

#print('SINR:',d.mean())
print('Data Rate:',d.mean())
```

```python
# For Elevation angles
# for Histograms




#d = df.loc[df.DL_Mode=='8PSK'].DL_capacity.to_numpy()
loaded_case = 'Cardiff'
d = df.loc[df.Ground_Station==loaded_case].Elevation_Angle.to_numpy()

 

fig, ax = plt.subplots()
n, bins, patches = plt.hist(x=d, bins='auto', color='#0504aa',
                            alpha=0.7, rwidth=0.85)
plt.grid(axis='y', alpha=0.75)
plt.xlabel('Angle (degrees)')
plt.ylabel('# of Occurence')
#plt.title(f'{link_type} Histogram of {loaded_case} SINR \n Margin {margin_input}')

link_type = 'A'
margin_input = '24 Hours'
plt.title(f'{link_type} Histogram of {loaded_case} Elevation Angles \n Margin {margin_input}')

 

print('Elevation angle:',d.mean())
```
